### Importing necessary modules 

```python
import pandas as pd
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
```

### Open Careerguide on Chrome

```python
browser = webdriver.Chrome()
browser.get("https://www.careerguide.com/career-options/")
```

### Search categories on careerguide

```python
category_list = []

categories = browser.find_elements(By.CLASS_NAME,"c-font-bold")
for category in range(1,len(categories)-12):
    category_list.append(categories[category].text)
    
print('{0}\n\nNumber of categories: {1}'.format(category_list, len(category_list)))
```

### Get sub categories from Categories 

```python
sub_categories = browser.find_elements(By.CLASS_NAME,"c-content-list-1")
sub_category_list = []
for sub_category in sub_categories:
    sub_category_list.append(sub_category.text.replace("\n",",").split(","))


sub_category_list
```

```python
len(sub_category_list)
```

```python
search_jobs = browser.find_element(By.ID,"jobs-search-box-keyword-id-ember24")
search_jobs_by_location = browser.find_element(By.ID,"jobs-search-box-location-id-ember24")
search_jobs.send_keys(category_list[0],Keys.ENTER)
```

### Job Search by category 

```python
searched_jobs = browser.find_elements(By.CLASS_NAME,"full-width.artdeco-entity-lockup__title.ember-view")
company_job1 = []

for job in searched_jobs:
    company_job1.append(job.text)
search_jobs.clear()

print('Job Category:',category_list[0])

```

### Job Offer Link

```python
job_link = browser.find_elements(By.ID,"ember61") 

job_link = job_link[0].get_attribute('href')
job_link
```

### Job States

```python
location = browser.find_elements(By.CLASS_NAME,"job-card-container__metadata-item")
location_name = []
for i in range(len(location)):
    location_name.append(location[i].text)

for y in location_name:
    if y == 'Remote' or y == 'On-site' or y == 'Hybrid':
        location_name.remove(y)

state_name = []
for i in location_name:
    loc = i.split(",")[1:-1]
    for q in loc:
        state_name.append(q[1:])

print('{0}\n\nNumber of state searches: {1}'.format(state_name, len(state_name)))

```

### Job search by subcategory and  state name 

```python
search_jobs_by_subcategory = browser.find_element(By.ID,"jobs-search-box-keyword-id-ember24")
search_jobs_by_subcategory.send_keys(sub_category_list[0][1])

search_jobs_by_state = browser.find_element(By.ID,"jobs-search-box-location-id-ember24")
search_jobs_by_state.send_keys(state_name[0],Keys.ENTER)
```

```python
# Company offering Job at

company_offering_job = browser.find_elements(By.CLASS_NAME,"job-card-container__metadata-item.job-card-container__metadata-item--workplace-type")
company_offering_job_list = []
for i in company_offering_job:
    company_offering_job_list.append(i.text)
    
print('{0}\n\nCount of companies offering jobs at: {1}'.format(company_offering_job_list, len(company_offering_job_list)))

```
